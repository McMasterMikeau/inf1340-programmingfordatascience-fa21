```python
#tidy data: what is it?
#tidy dataset has 3 properties:
# - each predictor/attribute/feature/variable forms a column and has well defined values
# - each observation/exemplar/instance forms a row
# - each observational unit forms a table/matrix/data frame
```

```python
#opposite of tidy dataset = untidy/messy data
#how do we prepare tidy data?
#there are 5 principles of tidy data

#Problem: Column names are values not variable names
#Principle: Column names need to be informative, variable names and not values
```

```python
#principle 5: the same observational unit should not be spread across multiple tables
#solution: join multiple tables/files into 1 singular data frame
#usual case: temporality aspect in the data

import glob
import re 

def extract_year(string):
    match = re.match(".+(\d{4})", string) 
    if match != None: return match.group(1)
    
path = './data'
allFiles = glob.glob(path + "/201*-baby-names-illinois.csv")
frame = pd.DataFrame()
df_list= []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    df.columns = map(str.lower, df.columns)
    df["year"] = extract_year(file_)
    df_list.append(df)
    
df = pd.concat(df_list)
df.head(20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generating fake data
data1 = np.random.normal(0,6,100)
data2 = np.random.normal(0,7,100)
data3 = np.random.normal(0,8,100)
data4 = np.random.normal(0,9,100)
data = list([data1,data2,data3,data4])
```

```python
#building boxplots
fig, ax = plt.subplots()

ax.boxplot(data)

#playing around with viz parameters
ax.set_title('box plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
xticklabels=['category 1','category 2','category 3','category 4']
ax.set_xticklabels(xticklabels)

ax.yaxis.grid(True)

plt.show()
```

```python
#violin plot

fig, ax = plt.subplots()

ax.violinplot(data, showmeans=False, showmedians = True)

ax.set_title('violin plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

xticklabels=['category 1','category 2','category 3','category 4']
ax.set_xticks([1,2,3,4])
ax.set_xticklabels(xticklabels)

ax.yaxis.grid(True)

plt.show()
```
